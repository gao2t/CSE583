```python
import base64
import pandas as pd
from IPython.display import HTML
def create_download_link( df2016, title = "Download CSV file", filename = "data2016.csv"):
    csv = df2016.to_csv()
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)

create_download_link(df2016)
```

```python
#plot the counts by hour
import matplotlib.pyplot as plt
import matplotlib as mpl
data2016 = pd.read_csv("/Users/tianyue/project/data/data2016.csv", index_col='Date', parse_dates=True)
data2016 = data2016.drop(columns = ['Unnamed: 0','The total (East + West) bicycle count','date','Hour of the day','The year'])
data2016
```
